```python
import subprocess

import random
import pprint
import re

import trivial
```

```python
arr_100 = [random.random() for i in range(100)]
arr_1000 = [random.random() for i in range(1000)]
arr_10000 = [random.random() for i in range(10000)]
arr_100000 = [random.random() for i in range(100000)]
```

```python
results = []
r = %timeit -o trivial.rolling_median_trivial(2, arr_100)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(4, arr_100)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(2, arr_1000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(4, arr_1000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(2, arr_10000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(4, arr_10000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(6, arr_10000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(8, arr_10000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(2, arr_100000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(4, arr_100000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(6, arr_100000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(8, arr_100000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(10, arr_100000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(12, arr_100000)
results.append(r)
r = %timeit -o trivial.rolling_median_trivial(14, arr_100000)
results.append(r)
```

```python
#best in microseconds
python_times = [r.best*1e3 for r in results]
```

```python
output = subprocess.check_output(["go.exe", "test", "-test.bench=.*",]).decode()
```

```python
pprint.pprint(output)
```

```python
golang_times = []
algo_times = []
for l in output.split('\n'):
    match = re.search('(.*)\s.*\s(\d*) ns/op', l)
    if match:
        method_name = match.group(1)
        if 'Algo' in method_name:
            algo_times.append(float(match.group(2)) / 1e6)
        else:
            golang_times.append(float(match.group(2)) / 1e6)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
index = [i for i in range(1, 1+len(python_times))]
index2 = [i+0.33 for i in index]
index3 = [i+0.66 for i in index]
print (index)
with plt.xkcd():
    plt.figure(figsize=(15, 8))
    plt.bar(index, python_times, 0.33, label='Python', color='orange', alpha=0.8)
    plt.bar(index2, golang_times, 0.33, label='Golang', color='red', alpha=0.8)
    plt.bar(index3, algo_times, 0.33, label='Golang', color='green', alpha=0.8)
    plt.ylabel('time in milliseconds')
    plt.legend(fontsize=20, loc='best')
```
